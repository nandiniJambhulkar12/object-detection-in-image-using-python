```python
 import matplotlib.pyplot as plt
```

```python
image = cv2.imread("coinsimage.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
blur = cv2.GaussianBlur(gray, (11,11), 0)
plt.imshow(blur, cmap="gray")
```

```python
canny= cv2.Canny(blur, 30, 150,3)
plt.imshow(canny, cmap="gray")
```

```python
(cnt.heirarchy)= cv2.findContours(canny.copy(), cv2.RETER_EXTERNAL, CV2.CHAIN_APPROX_NONE)
rgb= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt,-1, (0,255,0), 2)
plt.imshow(rgb)
```

```python
contours, _ = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
number_of_coins = len(contours)
```

```python
print("Number of coins:", number_of_coins)

# Draw contours on the original image for visualization
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# Display the original image with contours
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Detected Coins")
plt.show()
```

```python
print("Number of coins:", number_of_coins)
```

```python
import cv2
import cvlib as cv 
from cvlib.object_detection import draw_bbox
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("zebra.jpg")
box, label, c_score=cv.detect_common_objects(image)
output= draw_bbox(image,box,label, c_score)
plt.imshow(output)
plt.show()
print("no. of zebras:", label.count("zebra"))
```

```python
image = cv2.imread("zebra.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
blur = cv2.GaussianBlur(gray, (11,11), 0)
plt.imshow(blur, cmap="gray")
```

```python
canny= cv2.Canny(blur, 30, 150,3)
plt.imshow(canny, cmap="gray")
```

```python
contours, _ = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
number_of_zebras = len(contours)
```
